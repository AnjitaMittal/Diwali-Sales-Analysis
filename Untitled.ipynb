```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("Diwali Sales Data.csv", encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
df.shape
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount']=df['Amount'].astype(int)
```

```python
df.dtype()
```

```python
df['Amount'].dtype
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
gender_sales=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by=['Amount'], ascending=False)
```

```python
sns.barplot(x='Gender',y='Amount',data= gender_sales)
```

```python
sns.countplot(x='Age Group',data=df,hue='Gender')
```

```python
age_sales=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by=['Amount'], ascending=False)
age_sales
```

```python
sns.barplot(x='Age Group',y='Amount',data=age_sales)
```

```python
ax=sns.countplot(x='Marital_Status',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
maritalst_sales=df.groupby(['Marital_Status'],as_index=False)['Amount'].sum().sort_values(by=['Amount'], ascending=False)
maritalst_sales
```

```python
sns.barplot(data= maritalst_sales, x='Marital_Status',y='Amount')
```
